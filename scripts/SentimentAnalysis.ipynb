```python
linesDict = dict()

for person, screenLine in zip(df["Person"],df["ScreenLine"]):
    if person not in linesDict:
        linesDict[person] = ""
    try:
        linesDict[person] += " " + screenLine
    except TypeError:
        print(person, screenLine)
```

```python
import string, re
wnl = nltk.WordNetLemmatizer()

def wordTokenize(wordString: str):
    tokens = nltk.tokenize.word_tokenize(wordString)
    tokens = [i.lower() for i in tokens if i not in list(string.punctuation) and re.match(f'\w+',i)]
    lemmatizedTokens = [wnl.lemmatize(t) for t in tokens]
    freqdist = nltk.FreqDist(lemmatizedTokens)
    return tokens, lemmatizedTokens, freqdist
```

```python

freqdist = wordTokenize(linesDict["tyrion lannister"])[2]
print(sorted(freqdist.items(), key = lambda x: x[1]).reverse())
```
